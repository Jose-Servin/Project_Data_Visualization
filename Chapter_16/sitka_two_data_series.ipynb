```python
data = pd.read_csv('/Users/joseservin/Project_Data_Visualization/Chapter_16/sitka_weather_2014.csv')
data
```

# Find average max temperature grouped by month

```python
data.insert(1, 'Month', pd.DatetimeIndex(data['AKST']).month)
```

```python
month_data = data.groupby(['Month']).mean()
```

# Create February data DF

```python
data['AKST'] = pd.to_datetime(data['AKST'])
```

```python
feb_data = data.loc[data['Month'] == 2]
```

```python
len(feb_data.index) # 28 days in Feb 
```

```python
feb_data.dtypes
```

```python
x_axis = feb_data['AKST']
```

```python
y_axis = feb_data['Max TemperatureF']
```

```python
y_axis2 = feb_data['Min TemperatureF']
```

```python
plt.plot(x_axis, y_axis, c='red')
plt.plot(x_axis, y_axis2, c='blue')
plt.xticks(x_axis, rotation='vertical')
plt.fill_between(x_axis, y_axis, y_axis2, facecolor = 'blue', alpha = 0.1)
plt.show()
```
