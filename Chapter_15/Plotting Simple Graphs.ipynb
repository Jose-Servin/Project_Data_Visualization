# Plotting a simple line graph

```python
import matplotlib.pyplot as plt
squares  = [1,4,9,16,25] # squares of 1-5
plt.plot(squares)
plt.show()
```

## Changing the label type and graph thickness

```python
import matplotlib.pyplot as plt
squares  = [1,4,9,16,25] # squares of 1-5
plt.plot(squares,linewidth = 5)

# Set chart title and label axis 
plt.title("Square Numbers", fontsize = 24)
plt.xlabel("Value", fontsize  = 14)
plt.ylabel("Square of Value", fontsize = 14)

#Set size of tick labels
plt.tick_params(axis="both", labelsize = 14)

plt.show()
```

## Correcting the plot

```python
# plot() is assuming the first number in sequence corresponds to x-value 0
import matplotlib.pyplot as plt
input_values = [1,2,3,4,5] # giving input values to override the default behaviour 
squares  = [1,4,9,16,25] # squares of 1-5 (this represents our output values)
plt.plot(input_values, squares,linewidth = 5)

# Set chart title and label axis 
plt.title("Square Numbers", fontsize = 24)
plt.xlabel("Value", fontsize  = 14)
plt.ylabel("Square of Value", fontsize = 14)

#Set size of tick labels
plt.tick_params(axis="both", labelsize = 14)

plt.show()
```

## Plotting and styling individual points with scatter()

```python
plt.scatter(2,4, s=200) #providing an x and y value to plot; s sets the size of dot that represents each point

# Set chart title and label axes. 
plt.title("Square numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", which='major', labelsize=14)


plt.show()
```

## Plotting a Series of points with scater()

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

# Define scatter() parameters
plt.scatter(x_values,y_values, s=50)

# Set chart title and label axes. 
plt.title("Square numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", which='major', labelsize=14)

plt.show()
```

## Defining customer colors in your graph

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

# Define scatter() parameters
plt.scatter(x_values,y_values, c='red', s=50)

# Set chart title and label axes. 
plt.title("Square numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", which='major', labelsize=14)

plt.show()
```

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

# Define scatter() parameters
plt.scatter(x_values,y_values, c =(0,.5,0), s=50) # c values are between 0 and 1 (NOT RGB VALUES)

# Set chart title and label axes. 
plt.title("Square numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", which='major', labelsize=14)

plt.show()
```

## Using a colormap

```python
x_values = list(range(1,1001))
y_values = [x**2 for x in x_values] # the square of each value in the list x_values

plt.scatter(x_values, y_values, c = y_values, cmap = plt.cm.Blues, edgecolor = 'none', s = 12)

plt.title("Square numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", which='major', labelsize=14)

# Set the range for each axis (since we are dealing with larger numbers now)
plt.axis([0,1100,0,1100000])

plt.show()
```

# Calculating data automatically 

## Using lists and loops

```python
x_values = list(range(1,1001))
y_values = [x**2 for x in x_values] # the square of each value in the list x_values

plt.scatter(x_values,y_values,s=12)

plt.title("Square numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", which='major', labelsize=14)

# Set the range for each axis (since we are dealing with larger numbers now)
plt.axis([0,1100,0,1100000])

plt.show()
```
